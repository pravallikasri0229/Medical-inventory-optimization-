```python
total_returned_quantity = df["ReturnQuantity"].sum()
total_return_amount = (df["ReturnQuantity"] * df["Final_Sales"]).sum()

# Calculate overall return rate
total_sold_quantity = df["Quantity"].sum()
return_rate = total_returned_quantity / total_sold_quantity * 100

# Calculate average return amount per return transaction
average_return_per_transaction = total_return_amount / total_returned_quantity

# Print the calculated values
print("Total Returned Quantity:", total_returned_quantity)
print("Total Return Amount:", total_return_amount)
print("Return Rate (%):", return_rate)
print("Average Return Amount per Return Transaction:", average_return_per_transaction)
```
